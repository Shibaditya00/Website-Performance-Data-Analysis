# ***`Importing the dependencies`***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# ***`Reading the Dataset`***

```python
df = pd.read_csv("data-export (1).csv")
```

```python
df.head()
```

## **Note**
<pre>We observe that there is a proble in the columns section. The columns are named "Unnamed" which shouldn't be the case. The first row is the columns section. Hence, we need to remove the columns section and shift the first row to the columns section.</pre>

```python
df.columns = df.iloc[0]
df.head()
```

```python
df = df.drop(index = 0).reset_index(drop = True)
df.columns = ["Channel Group", 
              "DateHour", 
              "Users", 
              "Sessions", 
              "Engaged Sessions", 
              "Average engagement time per session", 
              "Engaged sessions per user", 
              "Events per session", 
              "Engagement rate", 
              "Event count"]
```

```python
df.head()
```

```python
df.info()
```

## **Note**
<pre>Here, we observe that the datatype of all the columns is mentioned as "object" whereas the datatype of "DateHour" column should be "datetime" and rest others should be in numeric form except "Channel Group" which is an "object" datatype.</pre>

```python
df['DateHour'] = pd.to_datetime(df["DateHour"], format = "%Y%m%d%H", errors = 'coerce')
```

```python
df.info()
```

```python
df.head()
```

```python
numeric_cols = df.columns.drop(["Channel Group", "DateHour"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors = 'coerce')
df["Hour"] = df["DateHour"].dt.hour
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# ***What patterns or trends can you observe in website sessions and users over time?***

```python
sns.set(style = 'whitegrid')
plt.figure(figsize = (10, 5))
df.groupby('DateHour')[["Sessions", "Users"]].sum().plot(ax = plt.gca())
plt.title("Sessions and Users over time")
plt.xlabel("DateHour")
plt.ylabel("count")
plt.show()
```

# ***Which marketing channel brought the highest number of users to the website? How can you use this insight to improve traffic from other sources?***

```python
plt.figure(figsize = (8, 5))
sns.barplot(data = df, x = "Channel Group", y = "Users", estimator = np.sum, palette = "viridis")
plt.title("Total Users By Channel")
plt.xticks(rotation = 45)
plt.show()
```

# ***Which channel has the highest average engagement time? What does that tell us about user behaviour and content effectiveness?***

```python
df.head()
```

```python
plt.figure(figsize = (8, 5))
sns.barplot(data = df, x = "Channel Group", y = "Average engagement time per session", estimator = np.mean, palette = "magma")
plt.title("Average Engagement Time By Channel")
plt.xticks(rotation = 45)
plt.show()
```

# ***How does engagement rate vary across different traffic channels?***

```python
plt.figure(figsize = (8, 5))
sns.boxplot(data = df, x = "Channel Group", y = "Engagement rate", palette = "coolwarm")
plt.title("Engagement Rate Distribution by Channel")
plt.xticks(rotation = 45)
plt.show()
```

# ***Which channels are driving more engagement sessions compared to non-engaged ones? Which startegies can improve engagement in underperforming channels?***

```python
df.head()
```

```python
session_df = df.groupby("Channel Group")[["Sessions", "Engaged Sessions"]].sum().reset_index()
session_df['Non-Engaged'] = session_df["Sessions"] - session_df["Engaged Sessions"]
session_df_melted = session_df.melt(id_vars = "Channel Group", value_vars = ["Engaged Sessions", "Non-Engaged"])

plt.figure(figsize = (8, 5))
sns.barplot(data = session_df_melted, x = "Channel Group", y = "value", hue = "variable")
plt.title("Engaged vs Non-Engaged Sessions")
plt.xticks(rotation = 45)
plt.show()
```

# ***At what hours of the day does each channle drive the most traffic?***

```python
heatmap_data = df.groupby(["Hour", "Channel Group"])["Sessions"].sum().unstack().fillna(0)

plt.figure(figsize = (12, 6))
sns.heatmap(heatmap_data, cmap = "YlGnBu", linewidths = .5, annot = True, fmt = '.0f')
plt.title("Traffic by Hour and Channel")
plt.xlabel("Channel Group")
plt.ylabel("Hour of Day")
plt.show()
```

# ***Is there any correlation between high traffic (sessions) and high engagement rate over time?***

```python
df_plot = df.groupby("DateHour")[["Engagement rate", "Sessions"]].mean().reset_index()

plt.figure(figsize = (10, 5))
plt.plot(df_plot["DateHour"], df_plot["Engagement rate"], label = "Engagement rate", color = "green")
plt.plot(df_plot["DateHour"], df_plot["Sessions"], label = "Sessions", color = "blue")
plt.title("Engagement Rate vs Sessions Over Time")
plt.xlabel("DateHour")
plt.legend()
plt.grid(True)
plt.show()
```
